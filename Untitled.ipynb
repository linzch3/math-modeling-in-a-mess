```matlab
seed = 1;rng(seed,'twister') % 设置随机种子
n = 500;rho = 0.15;p = 0.24;times = 500;v_limit = 5; % 设置模型参数
car_list = zeros(1,n);v = zeros(1,n);  % 建立车辆和车速矩阵
rand_sort = randperm(n);car_list(rand_sort(1:floor(n*rho))) = 1; % 产生车辆
road_store = zeros(times,n);road_store(1,:) = car_list; % 建立结果矩阵
for i = 2:times
    road_long = [car_list,car_list(1:find(car_list==1,1))];
    car_place_long = find(road_long==1);  % 车辆位置（首尾相接，方便计算）
    blank_num = diff(car_place_long) - 1; % 计算车辆前方的空单元格数
    car_place = find(car_list==1);  % 车辆
    car_v = v(car_place); % 获取车辆速度
    car_v = (car_v >= blank_num).*blank_num + ...% 减速、加速
            (car_v < blank_num & car_v < v_limit).*(car_v + 1) + ... % 加速
            (car_v < blank_num & car_v == v_limit).*car_v; % 速度不变
    car_v = car_v - (car_v>0).*(rand(1,length(car_v)) < p); % 随机慢化
    car_place_new = mod(car_place + car_v,n);  % 前进
    car_place_new(car_place_new == 0) = n; % 修正：位置0为n
    car_list = zeros(1,n); car_list(car_place_new) = 1; % 更新公路
    v = zeros(1,n); v(car_place_new) = car_v; % 更新速度
    road_store(i,:) = car_list; % 记录结果
end
[X,Y] = find(road_store);
plot(Y,X,'k.','MarkerSize',1);set(gca,'ydir','reverse')
xlabel('Distance');ylabel('Time');title(['Nagel-Schreckenbrg Traffic \rho=',num2str(rho),' p=',num2str(p)],'FontSize',16)
axis square
```
