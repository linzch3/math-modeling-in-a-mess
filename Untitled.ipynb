```matlab
seed = 1;rng(seed,'twister') % 
n = 500;rho = 0.15;p = 0.24;times = 500;v_limit = 5; % 
car_list = zeros(1,n);v = zeros(1,n);  % 
rand_sort = randperm(n);car_list(rand_sort(1:floor(n*rho))) = 1; % 
road_store = zeros(times,n);road_store(1,:) = car_list; % 
for i = 2:times
    road_long = [car_list,car_list(1:find(car_list==1,1))];
    car_place_long = find(road_long==1);  % 
    blank_num = diff(car_place_long) - 1; % 
    car_place = find(car_list==1);  % 
    car_v = v(car_place); % 
    car_v = (car_v >= blank_num).*blank_num + ...% 
            (car_v < blank_num & car_v < v_limit).*(car_v + 1) + ... % 
            (car_v < blank_num & car_v == v_limit).*car_v; % 
    car_v = car_v - (car_v>0).*(rand(1,length(car_v)) < p); % 
    car_place_new = mod(car_place + car_v,n);  % 
    car_place_new(car_place_new == 0) = n; % 0n
    car_list = zeros(1,n); car_list(car_place_new) = 1; % 
    v = zeros(1,n); v(car_place_new) = car_v; % 
    road_store(i,:) = car_list; % 
end
[X,Y] = find(road_store);
plot(Y,X,'k.','MarkerSize',1);set(gca,'ydir','reverse')
xlabel('Distance');ylabel('Time');title(['Nagel-Schreckenbrg Traffic \rho=',num2str(rho),' p=',num2str(p)],'FontSize',16)
axis square
```
